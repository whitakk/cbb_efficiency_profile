### Validate game_stats

```python
game_stats = pd.read_csv('../../data/game_stats.csv')
game_stats['date'] = pd.to_datetime(game_stats['date'])
```

Aggregates

```python
# most/fewest games per team
game_stats.groupby(['season','team']).date.count().reset_index().sort_values('date', ascending=False)
```

```python
# are non-d1 games included? 
# check if any opponents are not in team_list which is d1 only
game_stats['non_d1'] = ~game_stats['opp'].isin(game_stats.team.values)
game_stats['non_d1'].describe()
```

### Test merge matching

```python
game_stats_season = game_stats.groupby(['season','team']).date.count().reset_index()
game_stats_season.columns = ['season','team','games']

match_df = team_stats[['TeamName','season','eFG%']].merge(game_stats_season, left_on=['TeamName','season'], right_on=['team','season'], how='outer')
```

```python
match_df[pd.isnull(match_df.TeamName)]
```

```python
match_df[pd.isnull(match_df.team)]
```
